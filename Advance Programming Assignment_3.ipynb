#### **Assignment 1: Class Design and Encapsulation**

```python
# Task: Design a Student class for a university system.

class Student:
    
    # Constructor
    def __init__(self, student_id, name, program, gpa):
        self.student_id = student_id
        self.name = name
        self.program = program
        self.__gpa = 0.0   # private attribute
        
        # Use setter to validate GPA
        self.set_gpa(gpa)

    # Getter method
    def get_gpa(self):
        return self.__gpa

    # Setter method
    def set_gpa(self, gpa):
        if 0.0 <= gpa <= 4.0:
            self.__gpa = gpa
        else:
            print("Invalid GPA! GPA must be between 0.0 and 4.0")

    # Method to display student information
    def display_info(self):
        print("Student Information")
        print("Student ID:", self.student_id)
        print("Name:", self.name)
        print("Program:", self.program)
        print("GPA:", self.__gpa)


# Creating an object
student1 = Student("AAU001", "Dejene", "Bioinformatics", 3.8)

# Display information
student1.display_info()

# Access GPA using getter
print("\nCurrent GPA:", student1.get_gpa())

# Update GPA using setter
student1.set_gpa(3.9)

print("\nUpdated GPA:", student1.get_gpa())
```

##### **Concepts used**
Class → Student

Object → student1

Constructor → __init__()

Encapsulation → private GPA attribute __gpa

Getter and Setter → get_gpa() and set_gpa()

Validation → GPA allowed only between 0.0 and 4.0

#### **Assignment 2: Inheritance with Animal Classes**

```python
#Task: Create a parent class called Anima
class Animal:
    
    # Constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Parent method
    def speak(self):
        print("Animal makes a sound")


# Child class Dog
class Dog(Animal):

    # Overriding speak method
    def speak(self):
        print(self.name, "says: Woof Woof!")


# Child class Cow
class Cow(Animal):

    # Overriding speak method
    def speak(self):
        print(self.name, "says: Moo Moo!")


# Child class Bird
class Bird(Animal):

    # Overriding speak method
    def speak(self):
        print(self.name, "says: Chirp Chirp!")


# Creating objects
dog1 = Dog("Buddy", 3)
cow1 = Cow("Bella", 5)
bird1 = Bird("Sky", 1)

# Store objects in a list
animals = [dog1, cow1, bird1]

# Loop through the list and call speak()
for animal in animals:
    animal.speak()
```

##### **Concepts used**
Inheritance → Dog, Cow, and Bird inherit from Animal

Method Overriding → each child class has its own speak() method

Polymorphism → same method speak() behaves differently for different objects
                                                      
Objects and Lists → animal objects stored in a list and accessed using a loop

#### **Assignment 3: Bioinformatics Sequence System**

#Task: Build an OOP system for biological sequences.

```python
# Parent class
class Sequence:

    # Constructor
    def __init__(self, seq_id, sequence):
        self.seq_id = seq_id
        self.sequence = sequence.upper()

    # Method to calculate sequence length
    def get_length(self):
        return len(self.sequence)

    # Method to display information
    def display_info(self):
        print("Sequence ID:", self.seq_id)
        print("Sequence:", self.sequence)
        print("Length:", self.get_length())


# Child class for DNA
class DNASequence(Sequence):

    # Validate DNA sequence
    def validate(self):
        allowed = "ATGC"
        for base in self.sequence:
            if base not in allowed:
                return False
        return True

    # Calculate GC content
    def gc_content(self):
        g = self.sequence.count("G")
        c = self.sequence.count("C")
        gc = ((g + c) / len(self.sequence)) * 100
        return gc


# Child class for RNA
class RNASequence(Sequence):

    # Validate RNA sequence
    def validate(self):
        allowed = "AUGC"
        for base in self.sequence:
            if base not in allowed:
                return False
        return True

    # Convert RNA to DNA
    def to_dna(self):
        return self.sequence.replace("U", "T")


# Child class for Protein
class ProteinSequence(Sequence):

    # Validate protein sequence
    def validate(self):
        allowed = "ACDEFGHIKLMNPQRSTVWY"
        for aa in self.sequence:
            if aa not in allowed:
                return False
        return True

    # Count amino acids
    def amino_acid_count(self):
        counts = {}

        for aa in self.sequence:
            if aa in counts:
                counts[aa] += 1
            else:
                counts[aa] = 1

        return counts



# Creating objects


dna = DNASequence("DNA001", "ATGCGCTA")
rna = RNASequence("RNA001", "AUGCGCUA")
protein = ProteinSequence("PRO001", "MKTAFLL")


# DNA Operations

print("DNA Sequence")
dna.display_info()
print("Valid DNA:", dna.validate())
print("GC Content:", dna.gc_content(), "%")

print("\n------------------")


# RNA Operations

print("RNA Sequence")
rna.display_info()
print("Valid RNA:", rna.validate())
print("Converted DNA:", rna.to_dna())

print("\n------------------")


# Protein Operations

print("Protein Sequence")
protein.display_info()
print("Valid Protein:", protein.validate())
print("Amino Acid Count:", protein.amino_acid_count())
```

##### **Concepts used**
Inheritance → DNASequence, RNASequence, and ProteinSequence inherit from Sequence

Abstraction → common properties and methods are placed in the parent class

Scientific Modeling → represents biological sequences in an OOP system

Validation → checks allowed biological characters

Polymorphism → each child class has its own validation behavior

### **Assignment 4: Library Management System**
#Task: Design a library system using OOP.

```python
# Book class
class Book:

    # Constructor
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.available = True

    # Display book information
    def display_info(self):
        print(f"Book ID: {self.book_id}")
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Available: {self.available}")


# Member class
class Member:

    # Constructor
    def __init__(self, member_id, name):
        self.member_id = member_id
        self.name = name
        self.borrowed_books = []

    # Display member information
    def display_info(self):
        print(f"Member ID: {self.member_id}")
        print(f"Name: {self.name}")
        print("Borrowed Books:")

        if len(self.borrowed_books) == 0:
            print("No books borrowed")
        else:
            for book in self.borrowed_books:
                print("-", book.title)


# Library class
class Library:

    # Constructor
    def __init__(self, name):
        self.name = name
        self.books = []
        self.members = []

    # Add book to library
    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' added successfully.")

    # Register member
    def register_member(self, member):
        self.members.append(member)
        print(f"Member '{member.name}' registered successfully.")

    # Borrow book
    def borrow_book(self, member, book):

        if book.available:
            book.available = False
            member.borrowed_books.append(book)

            print(f"{member.name} borrowed '{book.title}'")

        else:
            print(f"Sorry! '{book.title}' is currently unavailable.")

    # Return book
    def return_book(self, member, book):

        if book in member.borrowed_books:
            member.borrowed_books.remove(book)
            book.available = True

            print(f"{member.name} returned '{book.title}'")

        else:
            print(f"{member.name} did not borrow '{book.title}'")



# Creating library


library = Library("AAU Library")

# Creating books
book1 = Book("B001", "Python Basics", "John Smith")
book2 = Book("B002", "Bioinformatics Guide", "Sarah Lee")

# Creating members
member1 = Member("M001", "Dejene")
member2 = Member("M002", "Abel")


# Add books and members


library.add_book(book1)
library.add_book(book2)

library.register_member(member1)
library.register_member(member2)

print("\n----------------------")


# Borrow books


library.borrow_book(member1, book1)

# Trying to borrow unavailable book
library.borrow_book(member2, book1)

print("\n----------------------")


# Display member information


member1.display_info()

print("\n----------------------")


# Return books


library.return_book(member1, book1)

# Borrow again after return
library.borrow_book(member2, book1)
```

##### **Concepts used**

Composition → Library contains books and members

Object Relationships → members borrow book objects from the library

State Management → book availability changes during borrowing and returning

Encapsulation → each class manages its own data and behavior

OOP Design → real-world library system modeled using classes and objects

### **Assignment 5: Abstract Base Class and Polymorphism**
#Task: Create an abstract class called Shape.

```python
from abc import ABC, abstractmethod
import math


# Abstract base class
class Shape(ABC):

    # Abstract method for area
    @abstractmethod
    def area(self):
        pass

    # Abstract method for perimeter
    @abstractmethod
    def perimeter(self):
        pass


# Child class Circle
class Circle(Shape):

    # Constructor
    def __init__(self, radius):
        self.radius = radius

    # Implement area method
    def area(self):
        return math.pi * self.radius ** 2

    # Implement perimeter method
    def perimeter(self):
        return 2 * math.pi * self.radius


# Child class Rectangle
class Rectangle(Shape):

    # Constructor
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Implement area method
    def area(self):
        return self.length * self.width

    # Implement perimeter method
    def perimeter(self):
        return 2 * (self.length + self.width)


# Child class Triangle
class Triangle(Shape):

    # Constructor
    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    # Implement area method using Heron's formula
    def area(self):
        s = (self.side1 + self.side2 + self.side3) / 2

        return math.sqrt(
            s * (s - self.side1) *
            (s - self.side2) *
            (s - self.side3)
        )

    # Implement perimeter method
    def perimeter(self):
        return self.side1 + self.side2 + self.side3



# Creating objects


circle = Circle(7)
rectangle = Rectangle(10, 5)
triangle = Triangle(3, 4, 5)

# Store all shapes in a list
shapes = [circle, rectangle, triangle]


# Using one loop for polymorphism


for shape in shapes:

    print("Shape:", shape.__class__.__name__)

    print("Area:", round(shape.area(), 2))

    print("Perimeter:", round(shape.perimeter(), 2))

    print("-------------------")
```

##### **Concepts used**

Abstraction → Shape is an abstract base class

Abstract Methods → area() and perimeter() must be implemented

Inheritance → child classes inherit from Shape

Polymorphism → same method calls work differently for each shape

Interface Design → all shapes follow the same structure and behavior

### **Assignment 6: Variant Annotation System**

#Task:Develop a simplified variant annotation program.

```python
# Parent class
class Variant:

    # Constructor
    def __init__(self, chromosome, position, ref_allele, alt_allele):
        self.chromosome = chromosome
        self.position = position
        self.ref_allele = ref_allele
        self.alt_allele = alt_allele

    # Method to classify variant
    def variant_type(self):
        return "General Variant"

    # Method to compute variant length
    def variant_length(self):
        return abs(len(self.alt_allele) - len(self.ref_allele))

    # Method to print annotation summary
    def annotation_summary(self):

        print("Chromosome:", self.chromosome)
        print("Position:", self.position)
        print("Reference Allele:", self.ref_allele)
        print("Alternate Allele:", self.alt_allele)
        print("Variant Type:", self.variant_type())
        print("Variant Length:", self.variant_length())


# Child class SNP
class SNP(Variant):

    # Override variant type
    def variant_type(self):
        return "Single Nucleotide Polymorphism (SNP)"

    # Override variant length
    def variant_length(self):
        return 1


# Child class Insertion
class Insertion(Variant):

    # Override variant type
    def variant_type(self):
        return "Insertion"

    # Override variant length
    def variant_length(self):
        return len(self.alt_allele) - len(self.ref_allele)


# Child class Deletion
class Deletion(Variant):

    # Override variant type
    def variant_type(self):
        return "Deletion"

    # Override variant length
    def variant_length(self):
        return len(self.ref_allele) - len(self.alt_allele)



# Creating objects


snp1 = SNP("Chr1", 1050, "A", "G")

ins1 = Insertion("Chr2", 2201, "T", "TAC")

del1 = Deletion("Chr3", 4500, "GCTA", "G")

# Store variants in a list
variants = [snp1, ins1, del1]


# Print annotation summaries


for variant in variants:

    print("Variant Annotation Summary")
    print("--------------------------")

    variant.annotation_summary()

    print()
```

##### **Concepts used**

Domain Modeling → models biological genetic variants

Inheritance → SNP, Insertion, and Deletion inherit from Variant

Method Overriding → each child class defines its own variant type and length calculation

Polymorphism → same method works differently for different variant objects

Scientific Data Representation → stores chromosome and allele information in an OOP structure

### **Assignment 7: Alternative Constructors**

#Task:Create a Genome class.

```python
# Genome class
class Genome:

    # Normal constructor
    def __init__(self, name, species, size):
        self.name = name
        self.species = species
        self.size = size

    # Class method to create object from CSV line
    @classmethod
    def from_csv_line(cls, csv_line):

        # Split the string using comma
        data = csv_line.split(",")

        name = data[0]
        species = data[1]
        size = int(data[2])

        return cls(name, species, size)

    # Class method to create object from dictionary
    @classmethod
    def from_dict(cls, data_dict):

        return cls(
            data_dict["name"],
            data_dict["species"],
            data_dict["size"]
        )

    # Method to display information
    def display_info(self):

        print("Genome Name:", self.name)
        print("Species:", self.species)
        print("Genome Size:", self.size, "Mb")



# Creating object using normal constructor


genome1 = Genome("Human", "Homo sapiens", 3200)


# Creating object using CSV line


csv_data = "NDama,Bos taurus,2800"

genome2 = Genome.from_csv_line(csv_data)


# Creating object using dictionary


genome_dict = {
    "name": "Rice",
    "species": "Oryza sativa",
    "size": 430
}

genome3 = Genome.from_dict(genome_dict)


# Display genome information

print("Genome 1")
genome1.display_info()

print("\nGenome 2")
genome2.display_info()

print("\nGenome 3")
genome3.display_info()
```

##### **Concepts used**

Constructors → normal constructor __init__()

Class Methods → from_csv_line() and from_dict()

Alternative Constructors → create objects using different data formats

Structured Data Parsing → converts CSV strings and dictionaries into objects

Object-Oriented Design → genome information represented using a class

### **Assignment 8: Static Methods vs Class Methods**

#Task: Create a DNAUtils class.

```python
# DNAUtils class
class DNAUtils:

    # Class variable
    valid_bases = "ATGC"

    # Static method to validate DNA sequence
    @staticmethod
    def is_valid_dna(sequence):

        sequence = sequence.upper()

        for base in sequence:
            if base not in DNAUtils.valid_bases:
                return False

        return True

    # Static method to calculate GC content
    @staticmethod
    def gc_content(sequence):

        sequence = sequence.upper()

        g = sequence.count("G")
        c = sequence.count("C")

        gc = ((g + c) / len(sequence)) * 100

        return gc

    # Class method to update valid bases
    @classmethod
    def update_valid_bases(cls, new_bases):

        cls.valid_bases = new_bases
        print("Updated valid bases:", cls.valid_bases)



# Testing the methods


sequence1 = "ATGCGCTA"

# Check if valid DNA
print("Is valid DNA:",
      DNAUtils.is_valid_dna(sequence1))

# Calculate GC content
print("GC Content:",
      DNAUtils.gc_content(sequence1), "%")

print("\nCurrent valid bases:",
      DNAUtils.valid_bases)

# Update valid bases
DNAUtils.update_valid_bases("ATGCU")

print("New valid bases:",
      DNAUtils.valid_bases)
```

##### **Concepts used**

Static Methods → methods that do not depend on object instances

Class Variables → shared variable valid_bases

Class Methods → modifies class-level data using cls

Utility Class Design → helper methods grouped in one class

Biological Sequence Validation → checks valid DNA characters and computes GC content

### **Assignment 9: Operator Overloading**

Task:Create a Sequence class.

```python
# Sequence class
class Sequence:

    # Constructor
    def __init__(self, seq_id, sequence):
        self.seq_id = seq_id
        self.sequence = sequence.upper()

    # Magic method for len()
    def __len__(self):
        return len(self.sequence)

    # Magic method for print()
    def __str__(self):
        return (f"Sequence ID: {self.seq_id}\n"
                f"Sequence: {self.sequence}\n"
                f"Length: {len(self)}")

    # Magic method for equality comparison
    def __eq__(self, other):
        return self.sequence == other.sequence

    # Magic method for addition
    def __add__(self, other):

        new_sequence = self.sequence + other.sequence

        return Sequence(
            self.seq_id + "_" + other.seq_id,
            new_sequence
        )



# Creating sequence objects


seq1 = Sequence("SEQ001", "ATGC")
seq2 = Sequence("SEQ002", "GCTA")
seq3 = Sequence("SEQ003", "ATGC")


# Testing __len__()

print("Length of seq1:", len(seq1))

print("\n-------------------")


# Testing __str__()


print(seq1)

print("\n-------------------")

# Testing __eq__()


print("seq1 == seq2 :", seq1 == seq2)

print("seq1 == seq3 :", seq1 == seq3)

print("\n-------------------")


# Testing __add__()


combined_seq = seq1 + seq2

print("Combined Sequence")
print(combined_seq)
```

##### **Concepts used**

Magic Methods → special methods like __len__(), __str__(), __eq__(), and __add__()

Operator Overloading → custom behavior for operators such as + and ==

Pythonic OOP → objects behave like built-in Python types

Object Comparison → compares biological sequences

Sequence Manipulation → joins two sequences using +
